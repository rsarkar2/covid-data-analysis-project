Importing libraries

```python
import pandas as pd
```

Reading enrichment dataset

```python
f1 = pd.read_csv("../../../../data/stage_1/political_leaning_fips.csv")
f1['county'] = f1['county'].str.lower()
f1['county'] = f1['county'].str.strip()
f1
```

Reading merged COVID-19 super dataset and making necessary changes to column names before merging

```python
f2 = pd.read_csv("../../../../data/stage_1/superdataset.csv")
f2.rename(columns = {'County_Name':'county'}, inplace = True)
f2.rename(columns = {'State':'state'}, inplace = True)
f2.rename(columns = {'countyFIPS':'countyfips'}, inplace = True)
f2
```

Merging the two dataframes to obtain the merged superdataset with COVID-19 data and presdiential election data. Data is merged using the 'countyfips' and 'state' columns.

```python
m1 = pd.merge(f2,f1,on=['countyfips','state'],how="inner")
m1
```

Saving merged file to a csv file

```python
m1.to_csv('../../../../data/stage_1/superdataset_political.csv',index=False)
```
